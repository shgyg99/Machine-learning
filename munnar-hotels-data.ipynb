This EDA is all about the hotels present in Munnar, Kerala. Munnar is a hill station which attracts large amount of tourists. This notebook explores the hotels present in this tourist destination

```python
import pandas as pd

data = pd.read_csv('/kaggle/input/hotels-in-munnar-kerala/munnar.csv')
```

```python
df = pd.DataFrame(data)
df2 = df.copy()
df2.head()
```

```python
df2 = df2.rename(columns={'Hotel Name\t\t\t\t\t\t\t\t\t':'name'})
```

```python
munnar = df2.drop(['Star Rating'], axis=1)
```

```python
munnar.head()
```

```python
for row in range(74):
    if ' m' in munnar['Distance to Landmark'][row]:
        munnar['Distance to Landmark'][row] = (float(munnar['Distance to Landmark'][row].split(' m')[0]))/1000.0
    elif ' km' in munnar['Distance to Landmark'][row]:
        munnar['Distance to Landmark'][row] = munnar['Distance to Landmark'][row].split(' km')[0]
```

```python
munnar['Distance to Landmark'] = munnar['Distance to Landmark'].astype(float)
munnar['Distance to Landmark'].describe()
```

```python
import numpy as np

munnar['Distance to Landmark'] = pd.cut(munnar['Distance to Landmark'],
                                   bins=[0, 5, 10, 15, 20, np.inf],
                                   labels=[1, 2, 3, 4, 5])
```

```python
munnar.head()
```

```python
munnar['Price'].unique()
```

```python
munnar['Price'].info()
```

```python
for i in range(74):
  if ',' in munnar['Price'][i]:
    a=munnar['Price'][i].split(',')
    b=f'{a[0]}{a[1]}'
    munnar['Price'][i]=float(b)
  else:
    munnar['Price'][i]=float(munnar['Price'][i])
```

```python
munnar.head()
```

```python
munnar['Price'] = munnar['Price'].astype(float)
```

```python
munnar['Price'].describe()
```

```python
munnar['Rating Description'].info
```

```python
munnar['Tax'].info()
```

```python
munnar['Tax'].unique()
```

```python
for i in range(74):
    if munnar['Tax'][i] == np.nan or type(munnar['Tax'][i]) == float or type(munnar['Tax'][i]) == int:
        continue
    elif ',' in munnar['Tax'][i]:
        a=munnar['Tax'][i].split(',')
        b=f'{a[0]}{a[1]}'
        munnar['Tax'][i]=float(b)
```

```python
munnar['Tax'] = munnar['Tax'].astype('float')
```

```python
munnar['Tax'] = munnar['Tax'].fillna(munnar['Tax'].mean())
```

```python
munnar['Tax'].isnull().sum()
```

```python
munnar['Tax'].unique()
```

```python
munnar['Tax'].describe()
```

```python
munnar.head()
```

```python
munnar['Nearest Landmark'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
munnar['Nearest Landmark'] = encoder.fit_transform(munnar[['Nearest Landmark']])
munnar['Location'] = encoder.fit_transform(munnar[['Location']])
```

```python
munnar['Rating Description'].unique()
```

```python
old = ['Excellent', 'Very Good', 'Good', 'Average']
new = [5, 4, 3, 2]
munnar['Rating Description'] = munnar['Rating Description'].replace(old, new)
```
